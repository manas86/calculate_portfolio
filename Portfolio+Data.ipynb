```python
import datetime as dt
symbols = ['AAPL', 'MMM', 'INTC', 'JPM', 'DE', 'WFC', 'BK', 'PM', 'HD', 'GE', 'GOOG', 'MSFT']
start_date = dt.date(2015,1,3)

#Benchmark Index
bench_symbol = "SPY"

#Dates
end_date = dt.date.today()
emo = end_date.month
eday = end_date.day
eyear = end_date.year
emonth = calendar.month_abbr[emo]

# print(end_date, start_date)
```

```python
# Update for Quandl
qsymbols = []
for i in range(len(symbols)):
    qsymbol = "WIKI/" + symbols[i].upper()
    qsymbols.append(qsymbol)

data = quandl.get(qsymbols, start_date=start_date, end_date=end_date, collapse="daily")
# print(data.head(10))

# extract only interested columns 
# this is for testing purpose but for specefic columns use below method
final_lists = [] 
for i in range(len(symbols)) :
        final_list = "WIKI/" + symbols[i].upper() + " - Close"
        final_lists.append(final_list)

# print(final_lists)
stocks = data[final_lists].copy()
stocks.columns = symbols
#print(stocks.head(10))
```

```python
# Update for Quandl 
qsymbols = []
for i in range(len(symbols)):
    qsymbol = "WIKI/" + symbols[i].upper() + ".4" #4th column is Close; Quandal columns start with 1
    qsymbols.append(qsymbol)

data = quandl.get(qsymbols, start_date=start_date, end_date=end_date, collapse="daily")
data.columns = symbols
data.head(10)
```

```python
import sys
import pandas as pd

# sanity check to make sure all the symbols are downloaded
temp_data = data.iloc[:,0:len(data.columns)].apply(pd.to_numeric)
for column in temp_data.columns:
    c = temp_data[column]
    if c.isnull().all():
        print('WARNING:  The following symbol: '+str(column)+' has no timeseries data. This could be due to an invalid ticker, or an entry not supported by Quandl. \n You will not be able to proceed with any function in the script until all of the symbols provided are downloaded.')
        sys.exit()
```

```python
allocations = [10, 10, 10, 10, 10, 10, 10, 10, 10, 10, 10, 10]
print(len(allocations)) # count of allocation
print(len(symbols)) # count of symbols
```

```python
port_val = data * allocations
```

```python
# Remove the rows with NO values
# calculate portfolio return value
port_val = port_val.fillna(port_val.mean())
port_val.head(10)
```

```python
port_val['Portfolio Value'] = port_val.sum(axis=1)
#port_val.head(10)
port_val.tail(1)
```

```python
# Calculate Portfolio Returns
port_rets = port_val.pct_change()
port_rets = port_rets.dropna(how='any')
```

```python
# calculate Portfolio Weights 
# check https://www.investopedia.com/terms/p/portfolio-weight.asp for general reference how to calculate weights 
# of a portfolio

assets = port_val.tail(1)
s = port_val.iloc[-1:, -1]
print(port_val.iloc[-1])
print(s)
port_weights = assets / int(s)
port_weights = port_weights.transpose()
port_weights.columns = ["Weight"]
port_weights = port_weights.drop(port_weights.index[len(port_weights) - 1])
```

```python
# Download portfolio data
root_path = ''
port_val.to_csv(root_path + '/Users/manassamantaray/Downloads/Portfolio Value.csv',index=True)
port_rets.to_csv(root_path + '/Users/manassamantaray/Downloads/Portfolio Returns.csv' ,index=True)
data.to_csv(root_path + '/Users/manassamantaray/Downloads/Portfolio Daily Prices.csv' ,index=True)
port_weights.to_csv(root_path + '/Users/manassamantaray/Downloads/Portfolio Weights.csv' ,index=True)

print('Portfolio data has successfully been downloaded.')
```
